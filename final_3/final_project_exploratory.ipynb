```python

dummy_workclass = pd.get_dummies(df['workclass'], prefix='workclass')
dummy_education = pd.get_dummies(df['education'], prefix='education')
dummy_marital_status = pd.get_dummies(df['marital.status'], prefix='marital.status')
dummy_occupation = pd.get_dummies(df['occupation'], prefix='occupation')
dummy_relationship = pd.get_dummies(df['relationship'], prefix='relationship')
dummy_race = pd.get_dummies(df['race'], prefix='race')
dummy_sex = pd.get_dummies(df['sex'], prefix='sex')
dummy_native_country = pd.get_dummies(df['native.country'], prefix='native.country')
```

```python
cols_to_keep = ['age', 'education.num', 'capital.loss','capital.gain','hours.per.week','income']
```
