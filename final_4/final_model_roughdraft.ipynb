```python
cols_to_keep = ['age', 'education.num', 'capital.loss','capital.gain','hours.per.week','income']
```

```python
handCalc = df[cols_to_keep].join(dummy_workclass.ix[:, 'workclass_Federal-gov':]).join(dummy_education.ix[:, 'education_11th':]).join(dummy_marital_status.ix[:, 'marital.status_Married-AF-spouse':]).join(dummy_occupation.ix[:, 'occupation_Adm-clerical':]).join(dummy_relationship.ix[:, 'relationship_Not-in-family':]).join(dummy_race.ix[:, 'race_Asian-Pac-Islander':]).join(dummy_sex.ix[:, 'sex_Male':]).join(dummy_native_country.ix[:, 'native.country_Cambodia':])
```

```python
handCalc.loc[handCalc['income'] == '<=50K','income'] = 0
```

```python
handCalc.loc[handCalc['income'] == '>50K', 'income'] = 1
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 500)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, model_train, y, scoring='roc_auc', cv=5)
print('CV AUC {}, Average AUC {}'.format(scores, scores.mean()))
```
